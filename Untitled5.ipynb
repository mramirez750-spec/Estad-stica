Para calcular el intervalo de confianza del 95%, seguiremos los siguientes pasos:

1.  Definir los parámetros dados.
2.  Calcular el valor crítico Z para un nivel de confianza del 95%.
3.  Calcular el margen de error.
4.  Construir el intervalo de confianza.

```python
import math
from scipy.stats import norm

# 1. Definir los parámetros dados
media_muestral = 170  # cm
desviacion_estandar_poblacional = 10  # cm
tamano_muestra = 50
nivel_confianza = 0.95

# 2. Calcular el valor crítico Z para un nivel de confianza del 95%
valor_critico_z = norm.ppf(1 - (1 - nivel_confianza) / 2)

# 3. Calcular el margen de error
margen_error = valor_critico_z * (desviacion_estandar_poblacional / math.sqrt(tamano_muestra))

# 4. Construir el intervalo de confianza
limite_inferior = media_muestral - margen_error
limite_superior = media_muestral + margen_error

print(f"Intervalo de confianza del {nivel_confianza*100}% para la media poblacional:")
print(f"[{limite_inferior:.4f}, {limite_superior:.4f}]")
```

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# 1. Definir las hipótesis nula y alternativa.
# H0: El peso promedio de los paquetes de cereal es 500 g (μ = 500)
# H1: El peso promedio de los paquetes de cereal es menor que 500 g (μ < 500) - Prueba de cola izquierda

# 2. Establecer el nivel de significancia α.
alpha = 0.05

# Parámetros dados del problema
mu_nula = 500  # Media poblacional bajo la hipótesis nula
media_muestral = 495  # Media de la muestra
desviacion_estandar_poblacional = 15  # Desviación estándar poblacional
tamano_muestra = 40

# 3. Identificar el estadístico de prueba (Z, t, χ²) y su distribución.
# Como conocemos la desviación estándar poblacional (σ) y el tamaño de la muestra es grande (n >= 30),
# utilizamos el estadístico de prueba Z y la distribución normal estándar.

# 4. Determinar el valor crítico y la región de rechazo.
# Para una prueba de cola izquierda con α = 0.05, el valor crítico Z es el valor tal que el área a su izquierda es α.
valor_critico_z = norm.ppf(alpha)

# La región de rechazo es Z < valor_critico_z

print(f"Nivel de significancia α: {alpha}")
print(f"Hipótesis Nula (H0): μ = {mu_nula} g")
print(f"Hipótesis Alternativa (H1): μ < {mu_nula} g")
print(f"Estadístico de prueba: Z (Distribución Normal Estándar)")
print(f"Valor crítico Z para α={alpha} (cola izquierda): {valor_critico_z:.4f}")
print(f"Región de rechazo: Z < {valor_critico_z:.4f}")


# 5. Calcule el estadístico de prueba y el valor p.

# Calcular el estadístico Z
estadistico_z = (media_muestral - mu_nula) / (desviacion_estandar_poblacional / np.sqrt(tamano_muestra))

# Calcular el valor p (área a la izquierda del estadístico_z en la distribución normal estándar)
valor_p = norm.cdf(estadistico_z)

print(f"\nEstadístico Z calculado: {estadistico_z:.4f}")
print(f"Valor p: {valor_p:.4f}")

# 6. Grafique la distribución correspondiente, indicando claramente la región de rechazo, el valor crítico y el estadístico calculado.

plt.figure(figsize=(10, 6))

# Rango de valores para la gráfica de la distribución normal estándar
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1) # PDF de la distribución normal estándar (media=0, desviación estándar=1)

plt.plot(x, y, label='Distribución Normal Estándar')

# Rellenar la región de rechazo (cola izquierda)
x_rechazo = np.linspace(-4, valor_critico_z, 100)
y_rechazo = norm.pdf(x_rechazo, 0, 1)
plt.fill_between(x_rechazo, y_rechazo, color='red', alpha=0.5, label='Región de Rechazo (α)')

# Marcar el valor crítico Z
plt.axvline(valor_critico_z, color='red', linestyle='--', label=f'Valor Crítico Z = {valor_critico_z:.4f}')

# Marcar el estadístico Z calculado
plt.axvline(estadistico_z, color='blue', linestyle='-', label=f'Estadístico Z Calculado = {estadistico_z:.4f}')

# Añadir texto para el valor p
plt.text(estadistico_z, 0.02, f'p-valor = {valor_p:.4f}', color='blue', ha='center')


plt.title('Distribución Normal Estándar con Región de Rechazo y Estadístico de Prueba')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tome una decisión estadística y concluya en términos prácticos, explicando el significado del valor p.
print("\nDecisión Estadística y Conclusión:")

if estadistico_z < valor_critico_z:
    decision_estadistica = "Rechazar H0"
    conclusion_practica = f"Con un nivel de significancia de {alpha}, hay suficiente evidencia para concluir que el peso promedio de los paquetes de cereal es significativamente menor que {mu_nula} g. Esto sugiere que la máquina podría estar descalibrada y subllenando los paquetes."
elif valor_p <= alpha:
     decision_estadistica = "Rechazar H0"
     conclusion_practica = f"Con un nivel de significancia de {alpha}, el valor p ({valor_p:.4f}) es menor o igual a α, lo que indica que hay suficiente evidencia para rechazar la hipótesis nula. Concluimos que el peso promedio de los paquetes de cereal es significativamente menor que {mu_nula} g, sugiriendo que la máquina podría estar descalibrada y subllenando."
else:
    decision_estadistica = "No rechazar H0"
    conclusion_practica = f"Con un nivel de significancia de {alpha}, no hay suficiente evidencia para rechazar la hipótesis nula. No podemos concluir que el peso promedio de los paquetes de cereal sea significativamente menor que {mu_nula} g. Los datos de la muestra no proporcionan evidencia sólida de que la máquina esté subllenando."

print(f"Decisión Estadística: {decision_estadistica}")
print(f"Conclusión Práctica: {conclusion_practica}")
print(f"\nSignificado del valor p ({valor_p:.4f}): Si la hipótesis nula (el peso promedio es {mu_nula} g) fuera verdadera, la probabilidad de obtener una media muestral tan baja como {media_muestral} g (o más baja) debido únicamente al azar es de aproximadamente {valor_p*100:.2f}%. Dado que este valor p ({valor_p:.4f}) es menor que el nivel de significancia α ({alpha}), consideramos que el resultado de la muestra es estadísticamente significativo y poco probable que ocurra por casualidad si la hipótesis nula fuera cierta.")
```

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# 1. Definir las hipótesis nula y alternativa.
# H0: El promedio de la clase es igual al promedio nacional (μ = 75)
# H1: El promedio de la clase es diferente al promedio nacional (μ ≠ 75) - Prueba de dos colas

# 2. Establecer el nivel de significancia α.
alpha = 0.05

# Parámetros dados del problema
mu_nula = 75  # Media poblacional bajo la hipótesis nula
media_muestral = 78  # Media de la muestra
desviacion_estandar_poblacional = 12  # Desviación estándar poblacional
tamano_muestra = 36

# 3. Identificar el estadístico de prueba (Z, t, χ²) y su distribución.
# Como conocemos la desviación estándar poblacional (σ) y el tamaño de la muestra es grande (n >= 30),
# utilizamos el estadístico de prueba Z y la distribución normal estándar.

# 4. Determinar los valores críticos y las regiones de rechazo.
# Para una prueba de dos colas con α = 0.05, dividimos α en dos (α/2 = 0.025) para cada cola.
# Los valores críticos Z son los valores tales que el área a su izquierda es α/2 y el área a su derecha es α/2.
valor_critico_z_inferior = norm.ppf(alpha / 2)
valor_critico_z_superior = norm.ppf(1 - alpha / 2)


# Las regiones de rechazo son Z < valor_critico_z_inferior o Z > valor_critico_z_superior

print(f"Nivel de significancia α: {alpha}")
print(f"Hipótesis Nula (H0): μ = {mu_nula} puntos")
print(f"Hipótesis Alternativa (H1): μ ≠ {mu_nula} puntos")
print(f"Estadístico de prueba: Z (Distribución Normal Estándar)")
print(f"Valores críticos Z para α={alpha} (dos colas): {valor_critico_z_inferior:.4f} y {valor_critico_z_superior:.4f}")
print(f"Regiones de rechazo: Z < {valor_critico_z_inferior:.4f} o Z > {valor_critico_z_superior:.4f}")


# 5. Calcule el estadístico de prueba y el valor p.

# Calcular el estadístico Z
estadistico_z = (media_muestral - mu_nula) / (desviacion_estandar_poblacional / np.sqrt(tamano_muestra))

# Calcular el valor p para una prueba de dos colas
# Es la suma de las áreas en las dos colas más allá del estadístico de prueba.
# Si estadistico_z es positivo, el valor p es 2 * P(Z > estadistico_z).
# Si estadistico_z es negativo, el valor p es 2 * P(Z < estadistico_z).
valor_p = 2 * (1 - norm.cdf(abs(estadistico_z)))

print(f"\nEstadístico Z calculado: {estadistico_z:.4f}")
print(f"Valor p: {valor_p:.4f}")

# 6. Grafique la distribución correspondiente, indicando claramente la región de rechazo, los valores críticos y el estadístico calculado.

plt.figure(figsize=(10, 6))

# Rango de valores para la gráfica de la distribución normal estándar
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1) # PDF de la distribución normal estándar (media=0, desviación estándar=1)

plt.plot(x, y, label='Distribución Normal Estándar')

# Rellenar la región de rechazo (dos colas)
x_rechazo_inferior = np.linspace(-4, valor_critico_z_inferior, 100)
y_rechazo_inferior = norm.pdf(x_rechazo_inferior, 0, 1)
plt.fill_between(x_rechazo_inferior, y_rechazo_inferior, color='red', alpha=0.5, label=f'Región de Rechazo (α/2 = {alpha/2})')

x_rechazo_superior = np.linspace(valor_critico_z_superior, 4, 100)
y_rechazo_superior = norm.pdf(x_rechazo_superior, 0, 1)
plt.fill_between(x_rechazo_superior, y_rechazo_superior, color='red', alpha=0.5) # No repetir la etiqueta


# Marcar los valores críticos Z
plt.axvline(valor_critico_z_inferior, color='red', linestyle='--', label=f'Valores Críticos Z = {valor_critico_z_inferior:.4f}, {valor_critico_z_superior:.4f}')
plt.axvline(valor_critico_z_superior, color='red', linestyle='--')


# Marcar el estadístico Z calculado
plt.axvline(estadistico_z, color='blue', linestyle='-', label=f'Estadístico Z Calculado = {estadistico_z:.4f}')

# Añadir texto para el valor p
plt.text(estadistico_z, 0.02, f'p-valor = {valor_p:.4f}', color='blue', ha='center')


plt.title('Distribución Normal Estándar con Regiones de Rechazo (Dos Colas) y Estadístico de Prueba')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tome una decisión estadística y concluya en términos prácticos, explicando el significado del valor p.
print("\nDecisión Estadística y Conclusión:")

if estadistico_z < valor_critico_z_inferior or estadistico_z > valor_critico_z_superior:
    decision_estadistica = "Rechazar H0"
    conclusion_practica = f"Con un nivel de significancia de {alpha}, hay suficiente evidencia para concluir que el promedio de la clase en el examen nacional ({media_muestral} puntos) es significativamente diferente del promedio nacional ({mu_nula} puntos). Los datos de la muestra respaldan la afirmación del profesor de que el promedio de su clase no es igual al promedio nacional."
elif valor_p <= alpha:
     decision_estadistica = "Rechazar H0"
     conclusion_practica = f"Con un nivel de significancia de {alpha}, el valor p ({valor_p:.4f}) es menor o igual a α, lo que indica que hay suficiente evidencia para rechazar la hipótesis nula. Concluimos que el promedio de la clase en el examen nacional es significativamente diferente del promedio nacional, respaldando la afirmación del profesor."
else:
    decision_estadistica = "No rechazar H0"
    conclusion_practica = f"Con un nivel de significancia de {alpha}, no hay suficiente evidencia para rechazar la hipótesis nula. No podemos concluir que el promedio de la clase en el examen nacional sea significativamente diferente del promedio nacional de {mu_nula} puntos. Los datos de la muestra no proporcionan evidencia sólida para respaldar la afirmación del profesor de que el promedio es diferente."

print(f"Decisión Estadística: {decision_estadistica}")
print(f"Conclusión Práctica: {conclusion_practica}")
print(f"\nSignificado del valor p ({valor_p:.4f}): Si la hipótesis nula (el promedio de la clase es {mu_nula} puntos) fuera verdadera, la probabilidad de obtener una media muestral tan extrema como {media_muestral} puntos (o más, en cualquier dirección) debido únicamente al azar es de aproximadamente {valor_p*100:.2f}%. Dado que este valor p ({valor_p:.4f}) es mayor que el nivel de significancia α ({alpha}), consideramos que el resultado de la muestra no es estadísticamente significativo al nivel α, y no rechazamos la hipótesis nula.")
```

```python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt

# 1. Definir las hipótesis nula y alternativa.
# H0: El promedio de la clase en matemáticas es 80 (μ = 80)
# H1: El promedio de la clase en matemáticas es diferente a 80 (μ ≠ 80) - Prueba de dos colas

# 2. Establecer el nivel de significancia α.
alpha = 0.01

# Parámetros dados del problema
mu_nula = 80  # Media poblacional bajo la hipótesis nula
media_muestral = 75  # Media de la muestra
desviacion_estandar_muestral = 8  # Desviación estándar muestral (s)
tamano_muestra = 12
grados_libertad = tamano_muestra - 1 # Grados de libertad para la distribución t

# 3. Identificar el estadístico de prueba (Z, t, χ²) y su distribución.
# Como no conocemos la desviación estándar poblacional (σ) y el tamaño de la muestra es pequeño (n < 30),
# utilizamos el estadístico de prueba t y la distribución t de Student.

# 4. Determine el valor crítico y la región de rechazo.
# Para una prueba de dos colas con α = 0.01, dividimos α en dos (α/2 = 0.005) para cada cola.
# Los valores críticos t son los valores tales que el área a su izquierda es α/2 y el área a su derecha es α/2.
valor_critico_t_inferior = t.ppf(alpha / 2, grados_libertad)
valor_critico_t_superior = t.ppf(1 - alpha / 2, grados_libertad)


# Las regiones de rechazo son t < valor_critico_t_inferior o t > valor_critico_t_superior

print(f"Nivel de significancia α: {alpha}")
print(f"Hipótesis Nula (H0): μ = {mu_nula} puntos")
print(f"Hipótesis Alternativa (H1): μ ≠ {mu_nula} puntos")
print(f"Estadístico de prueba: t (Distribución t de Student con {grados_libertad} grados de libertad)")
print(f"Valores críticos t para α={alpha} (dos colas): {valor_critico_t_inferior:.4f} y {valor_critico_t_superior:.4f}")
print(f"Regiones de rechazo: t < {valor_critico_t_inferior:.4f} o t > {valor_critico_t_superior:.4f}")


# 5. Calcule el estadístico de prueba y el valor p.

# Calcular el estadístico t
estadistico_t = (media_muestral - mu_nula) / (desviacion_estandar_muestral / np.sqrt(tamano_muestra))

# Calcular el valor p para una prueba de dos colas
# Es la suma de las áreas en las dos colas más allá del estadístico de prueba.
# Como la distribución t es simétrica, es 2 * P(t > |estadistico_t|)
valor_p = 2 * (1 - t.cdf(abs(estadistico_t), grados_libertad))


print(f"\nEstadístico t calculado: {estadistico_t:.4f}")
print(f"Valor p: {valor_p:.4f}")

# 6. Grafique la distribución correspondiente, indicando claramente la región de rechazo, los valores críticos y el estadístico calculado.

plt.figure(figsize=(10, 6))

# Rango de valores para la gráfica de la distribución t
# Ajustamos el rango para que sea relevante para la distribución t con los grados de libertad dados
x = np.linspace(t.ppf(0.0005, grados_libertad), t.ppf(0.9995, grados_libertad), 1000)
y = t.pdf(x, grados_libertad) # PDF de la distribución t

plt.plot(x, y, label=f'Distribución t de Student (gl={grados_libertad})')

# Rellenar la región de rechazo (dos colas)
x_rechazo_inferior = np.linspace(t.ppf(0.0005, grados_libertad), valor_critico_t_inferior, 100)
y_rechazo_inferior = t.pdf(x_rechazo_inferior, grados_libertad)
plt.fill_between(x_rechazo_inferior, y_rechazo_inferior, color='red', alpha=0.5, label=f'Región de Rechazo (α/2 = {alpha/2})')

x_rechazo_superior = np.linspace(valor_critico_t_superior, t.ppf(0.9995, grados_libertad), 100)
y_rechazo_superior = t.pdf(x_rechazo_superior, grados_libertad)
plt.fill_between(x_rechazo_superior, y_rechazo_superior, color='red', alpha=0.5) # No repetir la etiqueta


# Marcar los valores críticos t
plt.axvline(valor_critico_t_inferior, color='red', linestyle='--', label=f'Valores Críticos t = {valor_critico_t_inferior:.4f}, {valor_critico_t_superior:.4f}')
plt.axvline(valor_critico_t_superior, color='red', linestyle='--')


# Marcar el estadístico t calculado
plt.axvline(estadistico_t, color='blue', linestyle='-', label=f'Estadístico t Calculado = {estadistico_t:.4f}')

# Añadir texto para el valor p
plt.text(estadistico_t, 0.02, f'p-valor = {valor_p:.4f}', color='blue', ha='center')


plt.title(f'Distribución t de Student con Regiones de Rechazo (Dos Colas) y Estadístico de Prueba (gl={grados_libertad})')
plt.xlabel('Valor t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tome una decisión estadística y concluya en términos prácticos, explicando el significado del valor p.
print("\nDecisión Estadística y Conclusión:")

if abs(estadistico_t) > valor_critico_t_superior: # Para prueba de dos colas, comparamos el valor absoluto
    decision_estadistica = "Rechazar H0"
    conclusion_practica = f"Con un nivel de significancia de {alpha}, hay suficiente evidencia para concluir que el promedio de la clase en matemáticas ({media_muestral} puntos) es significativamente diferente del promedio afirmado ({mu_nula} puntos). Los datos de la muestra respaldan la idea de que el promedio de la clase no es 80."
elif valor_p <= alpha:
     decision_estadistica = "Rechazar H0"
     conclusion_practica = f"Con un nivel de significancia de {alpha}, el valor p ({valor_p:.4f}) es menor o igual a α, lo que indica que hay suficiente evidencia para rechazar la hipótesis nula. Concluimos que el promedio de la clase en matemáticas es significativamente diferente de {mu_nula} puntos, respaldando la idea de que el promedio no es 80."
else:
    decision_estadistica = "No rechazar H0"
    conclusion_practica = f"Con un nivel de significancia de {alpha}, no hay suficiente evidencia para rechazar la hipótesis nula. No podemos concluir que el promedio de la clase en matemáticas sea significativamente diferente de {mu_nula} puntos. Los datos de la muestra no proporcionan evidencia sólida para refutar la afirmación del colegio de que el promedio es 80."

print(f"Decisión Estadística: {decision_estadistica}")
print(f"Conclusión Práctica: {conclusion_practica}")
print(f"\nSignificado del valor p ({valor_p:.4f}): Si la hipótesis nula (el promedio es {mu_nula} puntos) fuera verdadera, la probabilidad de obtener una media muestral tan extrema como {media_muestral} puntos (o más, en cualquier dirección) debido únicamente al azar es de aproximadamente {valor_p*100:.2f}%. Dado que este valor p ({valor_p:.4f}) es mayor que el nivel de significancia α ({alpha}), consideramos que el resultado de la muestra no es estadísticamente significativo al nivel α, y no rechazamos la hipótesis nula.")
```

```python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt

# 1. Definir las hipótesis nula y alternativa.
# H0: La duración promedio de los focos es 10000 horas (μ = 10000)
# H1: La duración promedio de los focos es menor que 10000 horas (μ < 10000) - Prueba de cola izquierda

# 2. Establecer el nivel de significancia α.
alpha = 0.05

# Parámetros dados del problema
mu_nula = 10000  # Media poblacional bajo la hipótesis nula
media_muestral = 9800  # Media de la muestra
desviacion_estandar_muestral = 300  # Desviación estándar muestral (s)
tamano_muestra = 20
grados_libertad = tamano_muestra - 1 # Grados de libertad para la distribución t

# 3. Identificar el estadístico de prueba (Z, t, χ²) y su distribución.
# Como no conocemos la desviación estándar poblacional (σ) y el tamaño de la muestra es pequeño (n < 30),
# utilizamos el estadístico de prueba t y la distribución t de Student.

# 4. Determine el valor crítico y la región de rechazo.
# Para una prueba de cola izquierda con α = 0.05 y grados de libertad = tamano_muestra - 1,
# el valor crítico t es el valor tal que el área a su izquierda es α.
# La función ppf (percent point function) de scipy.stats.t devuelve el valor t para un cuantil dado.
# Para una prueba de cola izquierda, buscamos el valor t para α.
valor_critico_t = t.ppf(alpha, grados_libertad)

# La región de rechazo es t < valor_critico_t

print(f"Nivel de significancia α: {alpha}")
print(f"Hipótesis Nula (H0): μ = {mu_nula} horas")
print(f"Hipótesis Alternativa (H1): μ < {mu_nula} horas")
print(f"Estadístico de prueba: t (Distribución t de Student con {grados_libertad} grados de libertad)")
print(f"Valor crítico t para α={alpha} (cola izquierda): {valor_critico_t:.4f}")
print(f"Región de rechazo: t < {valor_critico_t:.4f}")


# 5. Calcule el estadístico de prueba y el valor p.

# Calcular el estadístico t
estadistico_t = (media_muestral - mu_nula) / (desviacion_estandar_muestral / np.sqrt(tamano_muestra))

# Calcular el valor p para una prueba de cola izquierda
# Es el área a la izquierda del estadistico_t en la distribución t de Student.
valor_p = t.cdf(estadistico_t, grados_libertad)


print(f"\nEstadístico t calculado: {estadistico_t:.4f}")
print(f"Valor p: {valor_p:.4f}")

# 6. Grafique la distribución correspondiente, indicando claramente la región de rechazo, el valor crítico y el estadístico calculado.

plt.figure(figsize=(10, 6))

# Rango de valores para la gráfica de la distribución t
# Ajustamos el rango para que sea relevante para la distribución t con los grados de libertad dados
x = np.linspace(t.ppf(0.001, grados_libertad), t.ppf(0.999, grados_libertad), 1000)
y = t.pdf(x, grados_libertad) # PDF de la distribución t

plt.plot(x, y, label=f'Distribución t de Student (gl={grados_libertad})')

# Rellenar la región de rechazo (cola izquierda)
x_rechazo = np.linspace(t.ppf(0.001, grados_libertad), valor_critico_t, 100)
y_rechazo = t.pdf(x_rechazo, grados_libertad)
plt.fill_between(x_rechazo, y_rechazo, color='red', alpha=0.5, label='Región de Rechazo (α)')

# Marcar el valor crítico t
plt.axvline(valor_critico_t, color='red', linestyle='--', label=f'Valor Crítico t = {valor_critico_t:.4f}')

# Marcar el estadístico t calculado
plt.axvline(estadistico_t, color='blue', linestyle='-', label=f'Estadístico t Calculado = {estadistico_t:.4f}')

# Añadir texto para el valor p
plt.text(estadistico_t, 0.02, f'p-valor = {valor_p:.4f}', color='blue', ha='center')


plt.title(f'Distribución t de Student con Región de Rechazo y Estadístico de Prueba (gl={grados_libertad})')
plt.xlabel('Valor t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

# 7. Tome una decisión estadística y concluya en términos prácticos, explicando el significado del valor p.
print("\nDecisión Estadística y Conclusión:")

if estadistico_t < valor_critico_t:
    decision_estadistica = "Rechazar H0"
    conclusion_practica = f"Con un nivel de significancia de {alpha}, hay suficiente evidencia para concluir que la duración promedio de los focos es significativamente menor que {mu_nula} horas. Esto sugiere que la afirmación del fabricante puede no ser precisa."
elif valor_p <= alpha:
     decision_estadistica = "Rechazar H0"
     conclusion_practica = f"Con un nivel de significancia de {alpha}, el valor p ({valor_p:.4f}) es menor o igual a α, lo que indica que hay suficiente evidencia para rechazar la hipótesis nula. Concluimos que la duración promedio de los focos es significativamente menor que {mu_nula} horas, sugiriendo que la afirmación del fabricante puede no ser precisa."
else:
    decision_estadistica = "No rechazar H0"
    conclusion_practica = f"Con un nivel de significancia de {alpha}, no hay suficiente evidencia para rechazar la hipótesis nula. No podemos concluir que la duración promedio de los focos sea significativamente menor que {mu_nula} horas. Los datos de la muestra no proporcionan evidencia sólida para refutar la afirmación del fabricante."

print(f"Decisión Estadística: {decision_estadistica}")
print(f"Conclusión Práctica: {conclusion_practica}")
print(f"\nSignificado del valor p ({valor_p:.4f}): Si la hipótesis nula (la duración promedio es {mu_nula} horas) fuera verdadera, la probabilidad de obtener una media muestral tan baja como {media_muestral} horas (o más baja) debido únicamente al azar es de aproximadamente {valor_p*100:.2f}%. Dado que este valor p ({valor_p:.4f}) es mayor que el nivel de significancia α ({alpha}), consideramos que el resultado de la muestra no es estadísticamente significativo al nivel α, y no rechazamos la hipótesis nula.")
```

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# 1. Definir las hipótesis nula y alternativa.
# H0: La varianza poblacional es 2.5 minutos² (σ² = 2.5)
# H1: La varianza poblacional difiere de 2.5 minutos² (σ² ≠ 2.5) - Prueba de dos colas

# 2. Establecer el nivel de significancia α.
alpha = 0.05

# Parámetros dados del problema
varianza_nula = 2.5  # Varianza poblacional bajo la hipótesis nula
varianza_muestral = 4  # Varianza de la muestra
tamano_muestra = 16
grados_libertad = tamano_muestra - 1 # Grados de libertad para la distribución chi-cuadrado

# 3. Identificar el estadístico de prueba (Z, t, χ²) y su distribución.
# Para pruebas de hipótesis sobre una varianza poblacional, utilizamos el estadístico de prueba χ²
# y la distribución chi-cuadrado con n-1 grados de libertad.

# 4. Determine los valores críticos y las regiones de rechazo.
# Para una prueba de dos colas con α = 0.05 y grados de libertad = tamano_muestra - 1,
# dividimos α en dos (α/2 = 0.025) para cada cola.
# Los valores críticos χ² son los valores tales que el área a su izquierda es α/2 y el área a su derecha es α/2.
valor_critico_chi2_inferior = chi2.ppf(alpha / 2, grados_libertad)
valor_critico_chi2_superior = chi2.ppf(1 - alpha / 2, grados_libertad)


# Las regiones de rechazo son χ² < valor_critico_chi2_inferior o χ² > valor_critico_chi2_superior

print(f"Nivel de significancia α: {alpha}")
print(f"Hipótesis Nula (H0): σ² = {varianza_nula} minutos²")
print(f"Hipótesis Alternativa (H1): σ² ≠ {varianza_nula} minutos²")
print(f"Estadístico de prueba: χ² (Distribución Chi-cuadrado con {grados_libertad} grados de libertad)")
print(f"Valores críticos χ² para α={alpha} (dos colas): {valor_critico_chi2_inferior:.4f} y {valor_critico_chi2_superior:.4f}")
print(f"Regiones de rechazo: χ² < {valor_critico_chi2_inferior:.4f} o χ² > {valor_critico_chi2_superior:.4f}")


# 5. Calcule el estadístico de prueba y el valor p.

# Calcular el estadístico χ²
estadistico_chi2 = ((tamano_muestra - 1) * varianza_muestral) / varianza_nula


# Calcular el valor p para una prueba de dos colas
# Es la suma de las áreas en las dos colas más allá del estadístico de prueba.
# Si estadistico_chi2 es menor que la media (gl), el valor p es 2 * P(χ² < estadistico_chi2).
# Si estadistico_chi2 es mayor que la media (gl), el valor p es 2 * P(χ² > estadistico_chi2).
# En general, es 2 * min(P(χ² <= estadistico_chi2), P(χ² >= estadistico_chi2))
valor_p = 2 * min(chi2.cdf(estadistico_chi2, grados_libertad), 1 - chi2.cdf(estadistico_chi2, grados_libertad))


print(f"\nEstadístico χ² calculado: {estadistico_chi2:.4f}")
print(f"Valor p: {valor_p:.4f}")

# 6. Grafique la distribución correspondiente, indicando claramente la región de rechazo, los valores críticos y el estadístico calculado.

plt.figure(figsize=(10, 6))

# Rango de valores para la gráfica de la distribución chi-cuadrado
# El rango debe comenzar en 0 y extenderse hasta un valor razonable para los grados de libertad dados.
x = np.linspace(0, chi2.ppf(0.999, grados_libertad), 1000)
y = chi2.pdf(x, grados_libertad) # PDF de la distribución chi-cuadrado

plt.plot(x, y, label=f'Distribución χ² (gl={grados_libertad})')

# Rellenar la región de rechazo (dos colas)
x_rechazo_inferior = np.linspace(0, valor_critico_chi2_inferior, 100)
y_rechazo_inferior = chi2.pdf(x_rechazo_inferior, grados_libertad)
plt.fill_between(x_rechazo_inferior, y_rechazo_inferior, color='red', alpha=0.5, label=f'Región de Rechazo (α/2 = {alpha/2})')

x_rechazo_superior = np.linspace(valor_critico_chi2_superior, chi2.ppf(0.999, grados_libertad), 100)
y_rechazo_superior = chi2.pdf(x_rechazo_superior, grados_libertad)
plt.fill_between(x_rechazo_superior, y_rechazo_superior, color='red', alpha=0.5) # No repetir la etiqueta


# Marcar los valores críticos χ²
plt.axvline(valor_critico_chi2_inferior, color='red', linestyle='--', label=f'Valores Críticos χ² = {valor_critico_chi2_inferior:.4f}, {valor_critico_chi2_superior:.4f}')
plt.axvline(valor_critico_chi2_superior, color='red', linestyle='--')


# Marcar el estadístico χ² calculado
plt.axvline(estadistico_chi2, color='blue', linestyle='-', label=f'Estadístico χ² Calculado = {estadistico_chi2:.4f}')

# Añadir texto para el valor p
# Ajustar la posición del texto del valor p para que no se superponga con el gráfico
plt.text(estadistico_chi2, chi2.pdf(estadistico_chi2, grados_libertad) + 0.01, f'p-valor = {valor_p:.4f}', color='blue', ha='center')


plt.title(f'Distribución χ² con Regiones de Rechazo (Dos Colas) y Estadístico de Prueba (gl={grados_libertad})')
plt.xlabel('Valor χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tome una decisión estadística y concluya en términos prácticos, explicando el significado del valor p.
print("\nDecisión Estadística y Conclusión:")

if estadistico_chi2 < valor_critico_chi2_inferior or estadistico_chi2 > valor_critico_chi2_superior:
    decision_estadistica = "Rechazar H0"
    conclusion_practica = f"Con un nivel de significancia de {alpha}, hay suficiente evidencia para concluir que la varianza poblacional del tiempo de duración de las llamadas es significativamente diferente de {varianza_nula} minutos². Los datos de la muestra respaldan la idea de que la varianza real no es 2.5."
elif valor_p <= alpha:
     decision_estadistica = "Rechazar H0"
     conclusion_practica = f"Con un nivel de significancia de {alpha}, el valor p ({valor_p:.4f}) es menor o igual a α, lo que indica que hay suficiente evidencia para rechazar la hipótesis nula. Concluimos que la varianza poblacional del tiempo de duración de las llamadas es significativamente diferente de {varianza_nula} minutos², respaldando la idea de que la varianza real no es 2.5."
else:
    decision_estadistica = "No rechazar H0"
    conclusion_practica = f"Con un nivel de significancia de {alpha}, no hay suficiente evidencia para rechazar la hipótesis nula. No podemos concluir que la varianza poblacional del tiempo de duración de las llamadas sea significativamente diferente de {varianza_nula} minutos². Los datos de la muestra no proporcionan evidencia sólida para refutar la afirmación de la compañía."

print(f"Decisión Estadística: {decision_estadistica}")
print(f"Conclusión Práctica: {conclusion_practica}")
print(f"\nSignificado del valor p ({valor_p:.4f}): Si la hipótesis nula (la varianza poblacional es {varianza_nula} minutos²) fuera verdadera, la probabilidad de obtener una varianza muestral tan extrema como {varianza_muestral} minutos² (o más, en cualquier dirección) debido únicamente al azar es de aproximadamente {valor_p*100:.2f}%. Dado que este valor p ({valor_p:.4f}) es mayor que el nivel de significancia α ({alpha}), consideramos que el resultado de la muestra no es estadísticamente significativo al nivel α, y no rechazamos la hipótesis nula.")
```

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# 1. Definir las hipótesis nula y alternativa.
# H0: La varianza poblacional es al menos 0.5 mm² (σ² >= 0.5)
# H1: La varianza poblacional es menor que 0.5 mm² (σ² < 0.5) - Prueba de cola izquierda

# 2. Establecer el nivel de significancia α.
alpha = 0.05

# Parámetros dados del problema
varianza_nula = 0.5  # Varianza poblacional bajo la hipótesis nula (valor límite de H0)
varianza_muestral = 0.3  # Varianza de la muestra
tamano_muestra = 12
grados_libertad = tamano_muestra - 1 # Grados de libertad para la distribución chi-cuadrado

# 3. Identificar el estadístico de prueba (Z, t, χ²) y su distribución.
# Para pruebas de hipótesis sobre una varianza poblacional, utilizamos el estadístico de prueba χ²
# y la distribución chi-cuadrado con n-1 grados de libertad.

# 4. Determine el valor crítico y la región de rechazo.
# Para una prueba de cola izquierda con α = 0.05 y grados de libertad = tamano_muestra - 1,
# el valor crítico χ² es el valor tal que el área a su izquierda es α.
# La función ppf (percent point function) de scipy.stats.chi2 devuelve el valor χ² para un cuantil dado.
# Para una prueba de cola izquierda, buscamos el valor χ² para α.
valor_critico_chi2 = chi2.ppf(alpha, grados_libertad)

# La región de rechazo es χ² < valor_critico_chi2

print(f"Nivel de significancia α: {alpha}")
print(f"Hipótesis Nula (H0): σ² >= {varianza_nula} mm²")
print(f"Hipótesis Alternativa (H1): σ² < {varianza_nula} mm²")
print(f"Estadístico de prueba: χ² (Distribución Chi-cuadrado con {grados_libertad} grados de libertad)")
print(f"Valor crítico χ² para α={alpha} (cola izquierda): {valor_critico_chi2:.4f}")
print(f"Región de rechazo: χ² < {valor_critico_chi2:.4f}")


# 5. Calcule el estadístico de prueba y el valor p.

# Calcular el estadístico χ²
estadistico_chi2 = ((tamano_muestra - 1) * varianza_muestral) / varianza_nula

# Calcular el valor p para una prueba de cola izquierda
# Es el área a la izquierda del estadistico_chi2 en la distribución chi-cuadrado.
valor_p = chi2.cdf(estadistico_chi2, grados_libertad)


print(f"\nEstadístico χ² calculado: {estadistico_chi2:.4f}")
print(f"Valor p: {valor_p:.4f}")

# 6. Grafique la distribución correspondiente, indicando claramente la región de rechazo, el valor crítico y el estadístico calculado.

plt.figure(figsize=(10, 6))

# Rango de valores para la gráfica de la distribución chi-cuadrado
# El rango debe comenzar en 0 y extenderse hasta un valor razonable para los grados de libertad dados.
x = np.linspace(0, chi2.ppf(0.999, grados_libertad), 1000)
y = chi2.pdf(x, grados_libertad) # PDF de la distribución chi-cuadrado

plt.plot(x, y, label=f'Distribución χ² (gl={grados_libertad})')

# Rellenar la región de rechazo (cola izquierda)
x_rechazo = np.linspace(0, valor_critico_chi2, 100)
y_rechazo = chi2.pdf(x_rechazo, grados_libertad)
plt.fill_between(x_rechazo, y_rechazo, color='red', alpha=0.5, label='Región de Rechazo (α)')

# Marcar el valor crítico χ²
plt.axvline(valor_critico_chi2, color='red', linestyle='--', label=f'Valor Crítico χ² = {valor_critico_chi2:.4f}')

# Marcar el estadístico χ² calculado
plt.axvline(estadistico_chi2, color='blue', linestyle='-', label=f'Estadístico χ² Calculado = {estadistico_chi2:.4f}')

# Añadir texto para el valor p
# Ajustar la posición del texto del valor p para que no se superponga con el gráfico
plt.text(estadistico_chi2, chi2.pdf(estadistico_chi2, grados_libertad) + 0.01, f'p-valor = {valor_p:.4f}', color='blue', ha='center')


plt.title(f'Distribución χ² con Región de Rechazo y Estadístico de Prueba (gl={grados_libertad})')
plt.xlabel('Valor χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

# 7. Tome una decisión estadística y concluya en términos prácticos, explicando el significado del valor p.
print("\nDecisión Estadística y Conclusión:")

if estadistico_chi2 < valor_critico_chi2:
    decision_estadistica = "Rechazar H0"
    conclusion_practica = f"Con un nivel de significancia de {alpha}, hay suficiente evidencia para concluir que la varianza poblacional en el grosor de las piezas metálicas es significativamente menor que {varianza_nula} mm². Esto sugiere que la varianza es menor que la garantizada por el control de calidad."
elif valor_p <= alpha:
     decision_estadistica = "Rechazar H0"
     conclusion_practica = f"Con un nivel de significancia de {alpha}, el valor p ({valor_p:.4f}) es menor o igual a α, lo que indica que hay suficiente evidencia para rechazar la hipótesis nula. Concluimos que la varianza poblacional es significativamente menor que {varianza_nula} mm², sugiriendo que es menor que la garantizada."
else:
    decision_estadistica = "No rechazar H0"
    conclusion_practica = f"Con un nivel de significancia de {alpha}, no hay suficiente evidencia para rechazar la hipótesis nula. No podemos concluir que la varianza poblacional en el grosor de las piezas metálicas sea significativamente menor que {varianza_nula} mm². Los datos de la muestra no proporcionan evidencia sólida para refutar la afirmación del control de calidad."

print(f"\nSignificado del valor p ({valor_p:.4f}): Si la hipótesis nula (la varianza poblacional es {varianza_nula} mm² o mayor) fuera verdadera, la probabilidad de obtener una varianza muestral tan baja como {varianza_muestral} mm² (o más baja) debido únicamente al azar es de aproximadamente {valor_p*100:.2f}%. Dado que este valor p ({valor_p:.4f}) es menor que el nivel de significancia α ({alpha}), consideramos que el resultado de la muestra es estadísticamente significativo y poco probable que ocurra por casualidad si la hipótesis nula fuera cierta.")
```
